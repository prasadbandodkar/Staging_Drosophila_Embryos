```python
# imports
import os
import cv2 as cv
import pandas as pd
from matplotlib import pyplot as plt
# %matplotlib inline
```

```python
# set up path information
#
foldername = '/Volumes/Extreme/Projects/staging'
path_data  = foldername + '/data/'
```

```python
# load images to be processed
#
fonum   = 3;
finum   = 3;
folders = [file[0] for file in os.walk(path_data)]
files   = os.listdir(folders[fonum])
file    = files[finum]
fileloc = folders[fonum] + '/' + file
print(fileloc)
```

```python
# Read image
#
image = cv.imread(fileloc, flags=cv.IMREAD_GRAYSCALE)
```

```python
plt.imshow(image, cmap='gray')
```
